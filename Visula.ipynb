```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Universities_Clustering.csv')
```

```python
df.iloc[0:len(df)-1,:]
```

```python
df.iloc[:]
```

```python
plt.hist(df['SAT'])
plt.xlabel ('SAT')
plt.ylabel('count')
plt.title('Histogram of SAT')
plt.show()
```

```python
plt.hist(df['Accept'])
plt.xlabel ('Accept')
plt.ylabel('count')
plt.title('Histogram of Accept')
plt.show()
```

```python
sns.histplot(df['Accept'])
plt.xlabel('Accept')
plt.ylabel('Values')
plt.title('Histogram of Accept')
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df['Accept'])
plt.show()
```

```python
plt.plot(df['Accept'],df['SAT'],'rs')
plt.xlabel('Accept')
plt.ylabel('SAT')
plt.title('Accept vs SAT')
plt.show()
```

```python
sns.pairplot(df)
```

```python
# histogram are univariable, scater plot are multi variable
sns.pairplot(df.iloc[:,1:3])
```
